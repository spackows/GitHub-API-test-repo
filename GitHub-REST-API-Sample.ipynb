## Use the Python `requests` library to call the GitHub REST API

About: `requests` library
- https://pypi.org/project/requests/
- https://www.w3schools.com/python/module_requests.asp

About: GitHub listing issues
- https://docs.github.com/en/rest/reference/issues#list-repository-issues

```python
import requests
url = "https://api.github.com/repos/spackows/GitHub-API-test-repo/issues"
response = requests.get( url )
```

## Dump results

About: `json` library
- https://docs.python.org/3/library/json.html
- https://www.w3schools.com/python/python_json.asp

```python
import json
results = response.json()
print( json.dumps( results, indent=3 ) )
```

## Sort issues

About: `sorted` function
- https://www.w3schools.com/python/ref_func_sorted.asp

About: `lambda` (annonymous) functions
- https://www.w3schools.com/python/python_lambda.asp

```python
sorted_results = sorted( results, key=lambda x: x["number"] )
print( json.dumps( sorted_results, indent=3 ) )
```

## Use a for-loop to print just the title and body

```python
for issue in sorted_results:
    print( issue["title"] + "\n" + issue["body"] + "\n\n---------------------------------------\n" )
```

## Listing all the comments for a given issue

About: GitHub listing comments for an issue
- https://docs.github.com/en/rest/reference/issues#list-issue-comments

```python
comments_url = "https://api.github.com/repos/spackows/GitHub-API-test-repo/issues/1/comments"
comments_response = requests.get( comments_url )
comments_results = comments_response.json()
print( json.dumps( comments_results, indent=3 ) )
```
